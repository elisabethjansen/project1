```python
from pathlib import Path
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a file path
census_data = pd.read_csv("UPDATED_census_data_2019_2023.csv")

# Read in the data. 
census_data.head()
```

```python
# Calculate owner-occupancy rate
census_data['owner_occupancy_rate'] = census_data['B25003_002E'] / census_data['B25003_001E']

# Group by year and calculate the mean owner-occupancy rate for each year
owner_occupancy_by_year = census_data.groupby('year')['owner_occupancy_rate'].mean()
```

```python
# Calculate the percentage of White householders
census_data['percent_white'] = census_data['B25006_002E'] / census_data['B01003_001E'] * 100

# Group by year and calculate the mean percentage for each year
percent_white_by_year = census_data.groupby('year')['percent_white'].mean()
```

```python
# Create a figure and a set of subplots
fig, ax = plt.subplots(2, 1, figsize=(10, 12))

# Plot owner-occupancy rate over time
owner_occupancy_by_year.plot(ax=ax[0], marker='o', linestyle='-')
ax[0].set_title('Owner-Occupancy Rate Over Time')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Owner-Occupancy Rate (%)')
ax[0].grid(True)

# Plot percentage of White householders over time
percent_white_by_year.plot(ax=ax[1], marker='o', color='red', linestyle='-')
ax[1].set_title('Percentage of White Householders Over Time')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Percentage (%)')
ax[1].grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
